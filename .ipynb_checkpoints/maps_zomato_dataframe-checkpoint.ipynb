```python
#import libraries

import pandas as pd
import numpy as np
```

```python
#read csv's:

maps_df= pd.read_csv("google_maps.csv")
zomato_df = pd.read_csv("lisbon_list_new.csv")
```

```python
#merging the 2 dataframes:

merged_df = pd.merge(maps_df, zomato_df, how='inner', left_on = 'name', right_on = 'name')
```

```python
merged_df["price_per_person"]= merged_df["price"] / 2
```
